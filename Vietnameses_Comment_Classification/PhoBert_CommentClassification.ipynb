```
maxlen = 125
embed_dim = 32  
num_heads = 2  
ff_dim = 32  
vocab_size=len(vocab)
epochs = 5
```

```
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import load_model
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.preprocessing.sequence import pad_sequences
from fairseq.data.encoders.fastbpe import fastBPE
from fairseq.data import Dictionary
import argparse
import re
import csv
from vncorenlp import VnCoreNLP
import pickle 
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```
from fairseq.data.encoders.fastbpe import fastBPE
from fairseq.data import Dictionary
bpe = fastBPE()
vocab = Dictionary()
vocab.add_from_file("/content/PhoBERT_base_transformers/dict.txt")
```

```
bpe.encode('Hôm nay, tôi là Vũ Tuấn Linh')
vocab.encode_line('<s> ' + 'Hôm nay, tôi là Vũ Tuấn Linh' + ' </s>')
```

```
train_csv = '/content/drive/MyDrive/Comment Classification/data/cleaned_train.csv'
```

```
def load_csv_data(filepath, textcol="text_ws"):
    df = pd.read_csv(filepath)
    samples = [ str(text) for text in df[textcol] ]
    labels  = [ int(intent) for intent in df["label"]]
    return samples, labels
```

```
train_data_text, train_data_label = load_csv_data(train_csv)
```

```
train_sents, test_sents, train_labels, test_labels = train_test_split(train_data_text, train_data_label, test_size=0.3)
```

```
train_ids = []
for sent in train_sents:
    subwords = '<s> ' + bpe.encode(sent) + ' </s>'
    encoded_sent = vocab.encode_line(subwords, append_eos=True, add_if_not_exist=False).long().tolist()
    train_ids.append(encoded_sent)

test_ids = []
for sent in test_sents:
    subwords = '<s> ' + bpe.encode(sent) + ' </s>'
    encoded_sent = vocab.encode_line(subwords, append_eos=True, add_if_not_exist=False).long().tolist()
    test_ids.append(encoded_sent)
    
train_ids = pad_sequences(train_ids, maxlen=maxlen, dtype="long", value=0, truncating="post", padding="post")
test_ids = pad_sequences(test_ids, maxlen=maxlen, dtype="long", value=0, truncating="post", padding="post")
```

```
from transformers import TFBertForSequenceClassification
phobert = TFBertForSequenceClassification.from_pretrained("vinai/phobert-base")
```

```
train_labels=np.array(train_labels)
test_labels=np.array(test_labels)
```

```
train_ids.shape
```

```
checkpoint_filepath = '/content/drive/MyDrive/Comment Classification/save_model/phoBERT_vietnamese_comment_classification_model'
model_checkpoint_callback = ModelCheckpoint (
    filepath=checkpoint_filepath,
    save_weights_only=True,
    monitor='val_accuracy',
    mode='max',
    verbose = 1,
    save_best_only=True
)
```

```
import tensorflow as tf
optimizer = tf.keras.optimizers.Adam(learning_rate=3e-5)
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
phobert.compile(optimizer=optimizer, loss=loss, metrics=["accuracy"])
H = phobert.fit(train_ids, train_labels,batch_size=64, epochs=5, validation_split = 0.3,callbacks=[model_checkpoint_callback])
```

```
import matplotlib.pyplot as plt
plt.plot(H.history['accuracy'])
plt.plot(H.history['val_accuracy'])
plt.title('model accuracy using phoBERT')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```
import matplotlib.pyplot as plt
plt.plot(H.history['loss'])
plt.plot(H.history['val_loss'])
plt.title('model loss using phoBERT')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```
score = phobert.evaluate(test_ids,test_labels)
print('Test loss: ', score[0])
print('Test accuracy: ', score[1])
```
